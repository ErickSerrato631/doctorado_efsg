```python
video_name = '/home/erick/Desktop/Doctorado/doctorado_efsg/new_model_10112024/weak_allee_mu_1_inprogres/fields_mu_1_weak.mp4'

# Obtener la lista de archivos de imagen en orden
images = [f'fields_block_1._step_{step:.3f}.png' for step in [0.005 * i for i in range(1, 200)]]
# /fields_block_1._step_0.995.png

# Crear el video
with imageio.get_writer(video_name, fps=10) as writer:
    for image_name in images:
        image_path = f'{image_folder}{image_name}'
        image = imageio.imread(image_path)
        writer.append_data(image)

print(f'Video guardado como {video_name}')
```
